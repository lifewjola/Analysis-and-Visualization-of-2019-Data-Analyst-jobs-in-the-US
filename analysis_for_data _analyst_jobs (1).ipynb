```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('2019_data_analyst_jobs.csv')
```

## How many Business Data Analysts got hired in New York in 2019?

```python
business_hired = len(df[(df['Hired'] == True) & (df['Job Location'] == 'New York, NY') & (df['Job Title'] == 'BUSINESS DATA ANALYST')])
print(f"{business_hired} Business Data Analysts got hired in New York in 2019")
```

## What percentage of Data Quality analyst jobs have easy apply?

```python
data_qual = len(df[df['Job Title'] == 'DATA QUALITY ANALYST'])
data_qual_easy_apply = len(df[(df['Job Title'] == 'DATA QUALITY ANALYST') & (df['Easy Apply'] == True)])
percent_data_qual_easy_apply = round(((data_qual_easy_apply / data_qual) * 100), 3)
print(f"{percent_data_qual_easy_apply}% of Data Quality Analyst Jobs have easy apply")
```

## What data analyst job is most easy to get? (Most hired)

```python
most_hired = df[df['Hired'] == True]['Job Title'].value_counts().idxmax()
print(f"{most_hired} is the easiest Data Analyst job to get")
```

## What month did data analysts (generally) get hired the most in 2019?

```python
month_most_hired = df[df['Hired'] == True]['Months'].value_counts().idxmax()
print(f"In 2019, Data Analysts got hired the most in {month_most_hired}")
```

## Data Quality Analysts were more in demand in what city in March 2019?

Find the city where Data Quality Analysts were hired the most in March

```python
city_most_hired_data_qual_march = df[(df['Job Title'] == 'DATA QUALITY ANALYST') & (df['Months'] == 'March') & (df['Hired'] == True)]['Job Location'].value_counts().idxmax()
print(f" Data Quality Analysts were more in demand in the city of {city_most_hired_data_qual_march} in March, 2019")
```

## What job title was the most demanded in the year 2019?

```python
job_most_demand = df[df['Hired'] == True]['Job Title'].value_counts().idxmax()
print(f"The most demanded Data Analyst Job in 2019 was: {job_most_demand}")
```

## How many Data Analysts get hired daily on average?

```python
# hired_per_day = mean(df[df['Hired' == True][

df['Date'] = pd.to_datetime(df['Date'])
hires_per_day = round(df[df['Hired'] == True]['Date'].value_counts().mean())

print(f"On average, {hires_per_day} Data Analyst(s) get hired daily")
```

## Which job title is the least competitive?
Job title that appeared least in the job title series

```python
least_competitive = df['Job Title'].value_counts().idxmin()
print(f"The least competitive Data Analyst job is: {least_competitive}")
```

## Which city has the least demand for data analysts? (Least hires)

```python
city_least_demand = df[df['Hired'] == True]['Job Location'].value_counts().idxmin()
print(f"The city with the least demand for data analysts is: {city_least_demand}")
```

## Show in a bar chart the number of hires for each data analyst job in 2019

```python
hire_per_job = df.groupby('Job Title')['Hired'].sum()
plt.figure(figsize=(60, 30))
plt.bar(hire_per_job.index, hire_per_job)
plt.xlabel('Job Title', fontsize=60)
plt.ylabel('Number of Hires', fontsize=60)
plt.title('Number of Hires per Job', fontsize=100)

plt.xticks(rotation=45, fontsize=40, ha='right')
plt.yticks(fontsize=40)
plt.show()
```

## A bar plot showing the hired rate of each job title by job location

```python
fig, axes = plt.subplots(2, 1, figsize=(20, 40))

dfh = df[df['Hired'] == True].groupby('Job Title').size().reset_index(name='Count')
sns.barplot(x='Count', y='Job Title', data=dfh, orient = 'h', ci=False, ax=axes[0])

dfg = df[df['Hired'] == False].groupby('Job Title').size().reset_index(name='Count')
sns.barplot(x='Count', y='Job Title', data=dfg, orient = 'h', ci=False, ax=axes[1])
```

```python
dfh = df[df['Hired'] == True].groupby(['Job Title', 'Months']).size().reset_index(name='Count')
u_hue = df['Months'].unique()
for hue_v in u_hue:
    df_hue = df[df['Months'] == hue_v]
    sns.barplot(x='Count', y='Job Title', data=dfh, orient = 'h', ci=False)
    plt.title(f'Report for {hue_v}')
    plt.xlabel('Number of Hires')
    plt.show()
```

```python
u_hue = df['Job Location'].unique()
num_plots = len(u_hue)
num_cols = 10 
num_rows = (num_plots + num_cols - 1) // num_cols  

# subplot
fig, axes = plt.subplots(num_rows, num_cols, figsize=(100, 100))

for idx, (hue_v, ax) in enumerate(zip(u_hue, axes.flatten())):
    df_hue = df[df['Job Location'] == hue_v]
    sns.barplot(x='Count', y='Job Title', data=dfh, orient = 'h', ci=False, ax=ax)
    plt.title(f'Number of Hires in {hue_v}')
    plt.xlabel('Number of Hires')

plt.tight_layout()
plt.show()
```
